DATASET USED : https://www.kaggle.com/datasets/canerbaloglu/ship-images-dataset

```python
import os
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, Flatten, Dense, Reshape
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import kagglehub

# Download dataset
path = kagglehub.dataset_download("canerbaloglu/ship-images-dataset")
print("Path to dataset files:", path)

# Load images
image_size = (128, 128)
valid_extensions = (".jpg", ".jpeg", ".png")
image_paths = []

if os.path.exists(path):
    for root, _, files in os.walk(path):
        for file in files:
            if file.lower().endswith(valid_extensions):
                image_paths.append(os.path.join(root, file))

# Reduce dataset size
num_images = 500  # Change this value to control the number of images used
image_paths = image_paths[:num_images]

def load_images(image_paths, size):
    images = []
    for img_path in image_paths:
        img = cv2.imread(img_path)
        if img is None:
            continue  # Skip if image is not loaded
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
        img = cv2.resize(img, size)
        img = img / 255.0  # Normalize to [0,1]
        images.append(img)
    images = np.array(images, dtype=np.float32)
    print(f"Loaded {len(images)} images with shape {images.shape}")
    return images

X = load_images(image_paths, image_size)

if len(X) == 0:
    raise ValueError("No valid image files found. Check dataset structure and ensure files are accessible.")

# Define vanilla autoencoder
input_shape = (128, 128, 3)
input_img = Input(shape=input_shape)

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = Conv2D(64, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
encoded = Dense(128, activation='relu')(x)  # Changed latent dimension to 64

# Decoder
d = Dense(256, activation='relu')(encoded)
d = Dense(32 * 32 * 128, activation='relu')(d)
d = Reshape((32, 32, 128))(d)
d = Conv2DTranspose(64, (3, 3), activation='relu', padding='same', strides=(2, 2))(d)
d = Conv2DTranspose(32, (3, 3), activation='relu', padding='same', strides=(2, 2))(d)
decoded = Conv2DTranspose(3, (3, 3), activation='sigmoid', padding='same')(d)

# Autoencoder model
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

autoencoder.summary()

# Ensure dataset is correctly shaped
if len(X.shape) != 4 or X.shape[1:] != (128, 128, 3):
    raise ValueError(f"Incorrect dataset shape: {X.shape}. Expected (num_samples, 128, 128, 3)")

# Train vanilla autoencoder
X_train = X[:int(0.8 * len(X))]
X_val = X[int(0.8 * len(X)):]

autoencoder.fit(X_train, X_train, validation_data=(X_val, X_val), epochs=20, batch_size=16)

# Display reconstructed images
def display_images(original, reconstructed, n=5):
    plt.figure(figsize=(10, 6))
    for i in range(n):
        # Original
        plt.subplot(2, n, i + 1)
        plt.imshow(original[i])
        plt.axis('off')

        # Reconstructed
        plt.subplot(2, n, i + 1 + n)
        plt.imshow(np.clip(reconstructed[i], 0., 1.))  # Ensure valid range
        plt.axis('off')
    plt.show()

reconstructed_imgs = autoencoder.predict(X_val[:5])
display_images(X_val[:5], reconstructed_imgs)
```
